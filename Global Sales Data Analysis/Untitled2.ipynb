```python
import pandas as pd
import numpy as np

# 1. Load the dataset (using unicode_escape to handle special characters)
df = pd.read_csv('/content/sales_data_sample.csv', encoding='unicode_escape')
```

```python
df.head(10)
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Display the new column names to verify
display(df.columns)
```

```python
#To find NUll Values.
df.isnull().sum()
```

```python
# Correct way to select multiple columns into a new DataFrame
newdf = df[['addressline2', 'state', 'postalcode', 'territory']]

# Display the first few rows of the new DataFrame
display(newdf.head())
```

```python
# 2. Handle Missing Values
# Drop ADDRESSLINE2 because it is 89% empty and unnecessary for business insights
df.drop(columns=['addressline2'], inplace=True)
```

```python
# Fill missing STATE and TERRITORY values so Power BI doesn't filter out those rows
df['state'].fillna('N/A', inplace=True)
df['territory'].fillna('Unknown', inplace=True)
```

```python
# Count NaN values (missing values represented by numpy.nan)
nan_count = df['postalcode'].isnull().sum()
print(f"Number of NaN values in 'postalcode': {nan_count}")

# Count empty strings (if any)
empty_string_count = (df['postalcode'] == '').sum()
print(f"Number of empty strings in 'postalcode': {empty_string_count}")

# Count 'N/A' strings (if any, case-insensitive)
na_string_count = df['postalcode'].astype(str).str.lower().eq('n/a').sum()
print(f"Number of 'N/A' strings in 'postalcode': {na_string_count}")

# Total missing/invalid entries
total_missing = nan_count + empty_string_count + na_string_count
print(f"Total missing/invalid entries in 'postalcode': {total_missing}")
```

```python
# Fill missing 'postalcode' values with 'N/A'
df['postalcode'].fillna('N/A', inplace=True)
```

```python
df.duplicated().sum()
```

```python
# Count the total number of duplicate rows
duplicate_rows_count = df.duplicated().sum()
print(f"Total number of duplicate rows: {duplicate_rows_count}")
```

```python
# Display the duplicate rows
# The 'keep=False' argument marks all occurrences of duplicates as True
display(df[df.duplicated(keep=False)])
```

```python
# 3. Fix Data Types
# Convert ORDERDATE from text to a proper datetime format for time-series analysis
df['orderdate'] = pd.to_datetime(df['orderdate'])
```

```python
# 4. Fix Inconsistencies
# Calculate the true 'Price Each' since the original column is capped at $100
df['actual_price'] = df['sales'] / df['quantityordered']
```
